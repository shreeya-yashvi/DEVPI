```python
#Q.11 calculate the confidence interval for the variables Quantity, Value and Weight.
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)  # Standard error of the mean
    margin_of_error = stderr * stats.t.ppf((1 + confidence) / 2, n - 1)  # t-distribution critical value
    return (mean - margin_of_error, mean + margin_of_error)

# Assuming 'data' is your dataframe
# Calculate confidence intervals for 'Quantity', 'Value', and 'Weight'
ci_quantity = confidence_interval(file_sample['Quantity'])
ci_value = confidence_interval(file_sample['Value'])
ci_weight = confidence_interval(file_sample['Weight'])

# Output the results
ci_results = {
    'Confidence Interval for Quantity': ci_quantity,
    'Confidence Interval for Value': ci_value,
    'Confidence Interval for Weight': ci_weight
}

print(ci_results)
```

The Confidence Interval for Value is between 4825.40 and 5078.34, reflecting stable transaction values.

MANAGERIAL INSIGHTS:

```python
#Q.17 Find the correlation between payment terms and countries? Does the selection of country affects the payment terms choices?(using Chi square test)
#taking:
#Null Hypothesis (H₀):
#There is no association between Payment Terms and Country. In other words, the choice of payment terms is independent of the customer's country.
#Alternative Hypothesis (H₁):
#There is an association between Payment Terms and Country. In other words, the choice of payment terms is dependent on the customer's country.

# Create a contingency table for 'Payment_Terms' and 'Country'
contingency_table = pd.crosstab(file_sample['Payment_Terms'], file_sample['Country'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print("")

#Determining Significance
significance_level= 0.05
if p< significance_level:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
print("")

print(expected)
```

FINDINGS:
P-Value:-
The p-value is 0.21756, which is greater than the 0.05 significance level. This means there is no statistically significant association between Payment Terms and Country in this dataset.
Conclusion: The observed differences in payment terms across different countries are likely due to random chance, and there is no strong evidence to suggest that the country a customer is from influences their payment terms.

Chi-Square Statistic:-
A Chi-Square statistic of 755.46 with 726 degrees of freedom suggests that, although there are some deviations from the expected values, they are not large enough to be statistically significant.
We fail to reject the Null Hypothesis meaning there is no significant association between Payment Terms and Country in this dataset.

MANAGERIAL INSIGHTS:
Based on the results of this Chi-Square test, country does not appear to have a significant impact on the choice of payment terms. This suggests that Payment Terms are likely set based on factors other than the customer's geographic location, such as company policy or transaction value.

Global Payment Flexibility: The lack of significant association suggests that customers across countries have similar preferences for payment terms. This allows businesses to streamline their payment policies globally, making it easier to manage accounts receivable without regional customizations.
